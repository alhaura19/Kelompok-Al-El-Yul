```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder   
```

```python
raw = np.loadtxt( 'heart.dat' )
raw 
```

```python
data_description = pd.read_csv('data_information.csv', names=['Name', 'Description'])
data = pd.DataFrame(raw)
data.columns = data_description['Name'].values
data.head(5)
```

```python
desclist = data_description['Name'].tolist()
datacat = {}
for i in range(len(desclist)):
    datacat[desclist[i]] = raw[:,i]
    
datacat
```

```python
bins = [0, 2, 18, 35, 65]
names_age = ['a0', 'a1', 'a2', 'a3', 'a4']

d = dict(enumerate(names_age, 1))
datacat['age'] = np.vectorize(d.get)(np.digitize(data['age'], bins))
datacat['age'][:5]
```

```python
datacat['sex'] = [str(datacat['sex'][i]) for i in range(len((datacat['sex'])))]
names_sex = ['male', 'female']

for i in range(len(datacat['sex'])):
    if datacat['sex'][i] == '1.0':
        datacat['sex'][i] = 'female'
    else:        
        datacat['sex'][i] = 'male'
        
datacat['sex'][:5]
```

```python
datacat['cp'] = [str(datacat['cp'][i]) for i in range(len(datacat['cp']))]

names_cp = ['cp1', 'cp2', 'cp3', 'cp4']
for i in range(len(datacat['cp'])):
    if datacat['cp'][i] == '1.0':
        datacat['cp'][i] = 'cp1'
    elif datacat['cp'][i] == '2.0':
        datacat['cp'][i] = 'cp2'
    elif datacat['cp'][i] == '3.0':
        datacat['cp'][i] = 'cp3'
    else:        
        datacat['cp'][i] = 'cp4'

datacat['cp'][:5]
```

```python
bins = [0, 120, 130, 140, 180]
names_rb = ['rb0', 'rb1', 'rb2', 'rb3', 'rb4']

d = dict(enumerate(names_rb, 1))

datacat['trestbps'] = np.vectorize(d.get)(np.digitize(datacat['trestbps'], bins))
datacat['trestbps'][:5]
```

```python
bins = [0, 200, 240]
names_sc = ['sc0', 'sc1', 'sc2']

d = dict(enumerate(names_sc, 1))

datacat['chol'] = np.vectorize(d.get)(np.digitize(datacat['chol'], bins))
datacat['chol'][:5]
```

```python
datacat['fbs'] = [str(datacat['fbs'][i]) for i in range(len(datacat['fbs']))]
names_fbs = ['fbs0', 'fbs1']

for i in range(len(datacat['fbs'])):
    if datacat['fbs'][i] == '1.0':
        datacat['fbs'][i] = 'fbs1'
    else:        
        datacat['fbs'][i] = 'fbs0'

datacat['fbs'][:5]
```

```python
datacat['restecg'] = [str(datacat['restecg'][i]) for i in range(len(datacat['restecg']))]
names_ekg = ['norm', 'abnormal', 'hyper']

for i in range(len(datacat['restecg'])):
    if datacat['restecg'][i] == '1.0':
        datacat['restecg'][i] = 'abnormal'
    elif datacat['restecg'][i] == '2.0':
        datacat['restecg'][i] = 'hyper'
    else:        
        datacat['restecg'][i] = 'norm'

datacat['restecg'][:5]
```

```python
bins = [0, 41, 106, 171, 236, 301]
names_mh = ['mh0', 'mh1', 'mh2', 'mh3', 'mh4', 'mh5']

d = dict(enumerate(names_mh, 1))

datacat['thalach'] = np.vectorize(d.get)(np.digitize(datacat['thalach'], bins))
datacat['thalach'][:5]
```

```python
datacat['exang'] = [str(datacat['exang'][i]) for i in range(len(datacat['exang']))]
names_ang = ['ang0','ang1']

for i in range(len(datacat['exang'])):
    if datacat['exang'][i] == '1.0':
        datacat['exang'][i] = 'ang1'
    else:        
        datacat['exang'][i] = 'ang0'

datacat['exang'][:5]
```

```python
bins = [0, 1, 2, 3, 4, 5, 6]
names_op = ['op0', 'op1', 'op2', 'op3', 'op4', 'op5', 'op6']

d = dict(enumerate(names_op, 1))

datacat['oldpeak'] = np.vectorize(d.get)(np.digitize(datacat['oldpeak'], bins))
datacat['oldpeak']
```

```python
datacat['slope'] = [str(datacat['slope'][i]) for i in range(len(datacat['slope']))]
names_sop = ['up', 'flat', 'down']

for i in range(len(datacat['slope'])):
    if datacat['slope'][i] == '1.0':
        datacat['slope'][i] = 'up'
    elif datacat['slope'][i] == '2.0':
        datacat['slope'][i] = 'flat'
    else:        
        datacat['slope'][i] = 'down'

datacat['slope'][:5]
```

```python
datacat['ca'] = [str(datacat['ca'][i]) for i in range(len(datacat['ca']))]
names_nmv = ['nmv0', 'nmv1', 'nmv2', 'nmv3']

for i in range(len(datacat['ca'])):
    if datacat['ca'][i] == '1.0':
        datacat['ca'][i] = 'nmv1'
    elif datacat['ca'][i] == '2.0':
        datacat['ca'][i] = 'nmv2'
    elif datacat['ca'][i] == '3.0':
        datacat['ca'][i] = 'nmv3'
    else:        
        datacat['ca'][i] = 'nmv0'

datacat['ca'][:5]
```

```python
datacat['thal'] = [str(datacat['thal'][i]) for i in range(len(datacat['thal']))]
names_thal = ['normal', 'fixed', 'reversible']

for i in range(len(datacat['thal'])):
    if datacat['thal'][i] == '3.0':
        datacat['thal'][i] = 'normal'
    elif datacat['thal'][i] == '6.0':
        datacat['thal'][i] = 'fixed'
    else:        
        datacat['thal'][i] = 'reversible'

datacat['thal'][:5]
```

```python
datacat['target'] = [str(datacat['target'][i]) for i in range(len(datacat['target']))]

for i in range(len(datacat['target'])):
    if datacat['target'][i] == '1.0':
        datacat['target'][i] = 'absence'
    else:        
        datacat['target'][i] = 'presence'

datacat['target'][:5]
```

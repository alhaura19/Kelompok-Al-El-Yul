## Data

```python
data = np.loadtxt( 'heart.dat' )
data
```

### Target

```python
target = data[:,13]
target
```

```python
# integer encode target
label_encoder = LabelEncoder()
target = label_encoder.fit_transform(target) # integer_encoded
target
```

### Fitur / Atribut untuk X

```python
atribute_labels = ['age', 'sex', 'chest_pain_type', 'resting_blood_pressure', 'serum_cholesterol_mg_per_dl', 'fasting_blood_sugar_gt_120_mg_per_dl', 'resting_ekg_results', 'max_heart_rate_achieved', 'exercise_induced_angina', 'oldpeak_eq_st_depression', 'slope_of_peak_exercise_st_segment', 'num_major_vessels', 'thal']
atribute_value = {}
for i in range(len(atribute_labels)):
    atribute_value[atribute_labels[i]] = data[:,i]
    
features = data[:,0:12]
# features
for j in range(len(atribute_labels)):
    print (atribute_labels[j])
    print (atribute_value[atribute_labels[j]])
```

```python
# integer encode number of major vessel
atribute_value['thal'] = label_encoder.fit_transform(atribute_value['thal'])
atribute_value['thal']
```

```python
# Age become category
bins = [0, 2, 18, 35, 65]
# names = ['<2', '2-18', '18-35', '35-65', '65+']
names = [0, 1, 2, 3, 4]

d = dict(enumerate(names, 1))

atribute_value['age'] = np.vectorize(d.get)(np.digitize(atribute_value['age'], bins))
atribute_value['age']
```

```python
# resting_blood_pressure become category
bins = [0, 120, 130, 140, 180]
# names = ['<120', '120-129', '130-139', '140-180', '180>']
# names = ['rb0', 'rb1', 'rb2', 'rb3', 'rb4']
names = [0, 1, 2, 3, 4]


d = dict(enumerate(names, 1))

atribute_value['resting_blood_pressure'] = np.vectorize(d.get)(np.digitize(atribute_value['resting_blood_pressure'], bins))
atribute_value['resting_blood_pressure']
```

```python
# serum_cholesterol_mg_per_dl become category
bins = [0, 200, 240]
# names = ['<200', '200-239', '240>']
# names = ['sc0', 'sc1', 'sc2']
names = [0, 1, 2]

d = dict(enumerate(names, 1))

atribute_value['serum_cholesterol_mg_per_dl'] = np.vectorize(d.get)(np.digitize(atribute_value['serum_cholesterol_mg_per_dl'], bins))
atribute_value['serum_cholesterol_mg_per_dl']
```

```python
# max_heart_rate_achieved become category
bins = [0, 41, 106, 171, 236, 301]
# names = ['<=40', '41-105', '106-170', '171-235', '236-300', '300>=']
# names = ['mh0', 'mh1', 'mh2', 'mh3', 'mh4', 'mh5']
names = [0, 1, 2, 3, 4, 5]

d = dict(enumerate(names, 1))

atribute_value['max_heart_rate_achieved'] = np.vectorize(d.get)(np.digitize(atribute_value['max_heart_rate_achieved'], bins))
atribute_value['max_heart_rate_achieved']
```

```python
# oldpeak_eq_st_depression become category
bins = [0, 1, 2, 3, 4, 5, 6]
# names = ['<1', '1-1.9', '2-2.9', '3-3.9', '4-4.9', '5-5.9', '6>']
# names = ['op0', 'op1', 'op2', 'op3', 'op4', 'op5', 'op6']
names = [0, 1, 2, 3, 4, 5, 6]

d = dict(enumerate(names, 1))

atribute_value['oldpeak_eq_st_depression'] = np.vectorize(d.get)(np.digitize(atribute_value['oldpeak_eq_st_depression'], bins))
atribute_value['oldpeak_eq_st_depression']
```

```python
for j in range(len(atribute_labels)):
    print (atribute_labels[j])
    print (atribute_value[atribute_labels[j]])
```

```python
# encode one hot coding all of atributes
for j in range(len(atribute_labels)):
    print (atribute_labels[j])
    encoded = to_categorical(atribute_value[atribute_labels[j]])
    print(encoded[:2])
    atribute_value[atribute_labels[j]] = encoded
```
